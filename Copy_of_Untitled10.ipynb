```python
#Process 2
import pandas as pd
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

data = {
    "text": [
        "Its raining outside",
        "I love to sleep",
        "At the same time I hate to go outside",
        "It would be better if I drink a coffee now",
        "I love family gathering"
    ]
}

df = pd.DataFrame(data)

df['score'] = df['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['sentiment'] = df['score'].apply(lambda x:
                                    'Positive' if x > 0
                                    else ('Negative' if x < 0
                                    else 'Neutral'))
df
```

```python
#process 4
#!pip install textblob
from textblob import TextBlob

sentences = ["Its raining outside", "I love to sleep", "At the same time I hate to go outside","I love family gathering"]
for s in sentences:
    blob = TextBlob(s)
    print(s, blob.sentiment)
```

Lexicon provides score with respect to sentiment type and that score is determined on the basis of sentiment intensity analyzer.while affin give only the score of sentiment .not mentions sentiment type.while the textblob provides the polarity which measures the positiveness or negativeness of a text and also the subjectivity of the text.
